```python
import plotly.express as px

df = pd.read_csv("IMDB-Movie-Data.csv")
df.columns = df.columns.str.strip()

df_genres = df.assign(Genre=df['Genre'].str.split(',')).explode('Genre')
df_genres['Genre'] = df_genres['Genre'].str.strip()

avg_rating = df_genres.groupby(['Genre', 'Director'])['Rating'].mean().reset_index()

custom_colors = [
    [0.0, "pink"],
    [0.5, "purple"],
    [1.0, "blue"]
]

fig = px.sunburst(
    avg_rating,
    path=['Genre', 'Director'],
    values='Rating',
    color='Rating',
    color_continuous_scale=custom_colors,
    title="Sunburst of Average Movie Ratings by Genre and Director"
)
fig.show()
```

```python
import pandas as pd
import networkx as nx
import plotly.graph_objects as go

df = pd.read_csv("IMDB-Movie-Data.csv")
df.columns = df.columns.str.strip()
df_top = df.sort_values(by='Revenue (Millions)', ascending=False).head(100)
df_top['Actors'] = df_top['Actors'].str.split(',')
df_top = df_top.explode('Actors')
df_top['Actors'] = df_top['Actors'].str.strip()
from itertools import combinations
edges = []
for actors in df_top.groupby('Title')['Actors'].apply(list):
    if len(actors) > 1:
        edges.extend(combinations(actors, 2))
G = nx.Graph()
G.add_edges_from(edges)
plt.figure(figsize=(12,12))
nx.draw_circular(G, with_labels=True, node_color='pink', edge_color='purple', node_size=400, font_size=8)
plt.title("Actor Collaboration Circular Network")
plt.show()
```

```python
import plotly.express as px
df_clean = df.dropna(subset=['Rating', 'Revenue (Millions)', 'Runtime (Minutes)', 'Votes'])
fig = px.parallel_coordinates(
    df_clean,
    dimensions=['Rating','Revenue (Millions)','Runtime (Minutes)','Votes'],
    color='Rating',
    color_continuous_scale=px.colors.sequential.Pinkyl,
    title="Parallel Coordinates: Multi-dimensional Movie Analysis"
)
fig.show()
```
